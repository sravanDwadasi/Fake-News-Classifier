About the Dataset:

1. id: unique id for a news article
2. title: the title of a news article
3. author: author of the news article
4. text: the text of the article; could be incomplete

```python
# replacing the null values with empty string
news_dataset = news_dataset.fillna('')
```

```python
# separating the data & label
X = news_dataset.drop(columns='label', axis=1)
Y = news_dataset['label']
```

```python
print(X)
print(Y)
```

Stemming:

Stemming is the process of reducing a word to its Root word

example:
actor, actress, acting --> act

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(stemming)
```

```python
print(news_dataset['content'])
```

```python
#separating the data and label
X = news_dataset['content'].values
Y = news_dataset['label'].values
```

```python
print(X)
```

```python
print(Y)
```

```python
Y.shape
```

```python
# converting the textual data to numerical data
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
print(X)
```

### Logistic Regression

```python
X_train_log, X_test_log, Y_train_log, Y_test_log = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
model1 = LogisticRegression()
```

```python
model1 = model1.fit(X_train_log, Y_train_log)
```

```python
# accuracy score on the training data
X_train_prediction_log = model1.predict(X_train_log)
training_data_accuracy_log = accuracy_score(X_train_prediction_log, Y_train_log)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy_log)
```

```python
# accuracy score on the test data
X_test_prediction_log = model1.predict(X_test_log)
test_data_accuracy_log = accuracy_score(X_test_prediction_log, Y_test_log)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy_log)
```

Classifying through test data

```python
X_new_log = X_test_log[2]

prediction_log = model1.predict(X_new_log)
print(prediction_log)

if (prediction_log[0]==0):
  print('The news is Real')
else:
  print('The news is Fake')
```
